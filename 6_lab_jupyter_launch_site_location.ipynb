```python
%pip install geopy

import pandas as pd
from geopy.distance import geodesic

# Crear un dataframe con las coordenadas de los sitios de lanzamiento
launch_sites = pd.DataFrame({
    'Launch Site': ['Site A', 'Site B', 'Site C', 'Site D'],
    'Lat': [28.562302, 34.632834, 27.951267, 25.997264],
    'Long': [-80.577356, -120.610746, -80.558074, -97.156546]
})

# Crear una funcin para calcular la distancia entre dos puntos geogrficos
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).km

# Calcular las distancias entre un sitio de lanzamiento y todos los dems sitios
distances = []
for index, row in launch_sites.iterrows():
    launch_site_coord = (row['Lat'], row['Long'])
    for index2, row2 in launch_sites.iterrows():
        if index != index2:
            proximity_coord = (row2['Lat'], row2['Long'])
            distance = calculate_distance(launch_site_coord, proximity_coord)
            distances.append({
                'Launch Site': row['Launch Site'],
                'Proximity Site': row2['Launch Site'],
                'Distance (km)': distance
            })

# Crear un dataframe con las distancias calculadas
distances_df = pd.DataFrame(distances)

# Mostrar el dataframe de distancias
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Ejemplo de coordenadas de un sitio de lanzamiento y un punto de inters
launch_site_coord = (28.562302, -80.577356)  # Coordenadas del sitio de lanzamiento
railway_coord = (28.563302, -80.578356)      # Coordenadas del ferrocarril (ejemplo)

# Calcular la distancia entre el sitio de lanzamiento y el ferrocarril
distance = calculate_distance(launch_site_coord[0], launch_site_coord[1], railway_coord[0], railway_coord[1])

print(f"La distancia entre el sitio de lanzamiento y el ferrocarril es de {distance:.2f} km")
```
